```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "train.csv"  # Change this to your actual file path
df_titanic = pd.read_csv(file_path)

# Check for missing values
print("Missing Values:\n", df_titanic.isnull().sum())

# Summary statistics
print("\nSummary Statistics:\n", df_titanic.describe())

# Data Cleaning
df_titanic['Age'] = df_titanic['Age'].fillna(df_titanic['Age'].median())  # Fill missing age with median
df_titanic['Embarked'] = df_titanic['Embarked'].fillna(df_titanic['Embarked'].mode()[0])  # Fill missing Embarked with mode
df_titanic.drop(columns=['Cabin'], inplace=True)  # Drop Cabin due to excessive missing values

# Convert categorical variables
df_titanic['Sex'] = df_titanic['Sex'].map({'male': 0, 'female': 1})
df_titanic['Embarked'] = df_titanic['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})

# Visualization 1: Survival Rate by Passenger Class
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', hue='Pclass', data=df_titanic, errorbar=None, palette="Blues", legend=False)
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Passenger Class")
plt.show()

# Visualization 2: Survival Rate by Gender
plt.figure(figsize=(6, 4))
sns.barplot(x='Sex', y='Survived', data=df_titanic, errorbar=None, palette="coolwarm")
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender")
plt.show()

# Visualization 3: Age Distribution by Survival
plt.figure(figsize=(10, 6))
sns.histplot(df_titanic, x="Age", hue="Survived", bins=20, kde=True, palette="Set1", alpha=0.6)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution by Survival")
plt.legend(["Did not survive", "Survived"])
plt.show()

# Visualization: Survival Rate by Gender (Fixed)
import seaborn as sns
import matplotlib.pyplot as plt

# Visualization: Survival Rate by Gender (Fixed)
plt.figure(figsize=(6, 4))
sns.barplot(x='Sex', y='Survived',  hue='Sex',  data=df_titanic, errorbar=None, palette="coolwarm", dodge=False)
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender")
plt.show()

```
